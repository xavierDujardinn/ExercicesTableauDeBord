```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('ChiffresVente.csv', sep=';')
df
```

```python
width = 0.17
x = np.arange(len(df['Mois']))

plt.figure(figsize=(14, 6))
plt.bar(x - width, df['2009'], width, label='2009', color='#4A90E2')
plt.bar(x,         df['2010'], width, label='2010', color='#FF6B6B')
plt.bar(x + width, df['2011'], width, label='2011', color='#6AB04C')
plt.legend()
plt.xticks(x, df['Mois'])
plt.xlabel('Mois')
plt.ylabel('Chiffre de vente')
plt.title('Chiffre de vente par mois pour chaque année entre 2009 et 2011')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df['Mois'], df['2009'], label='2009', color='#4A90E2', marker='o', linestyle='--')
plt.plot(df['Mois'], df['2010'], label='2010', color='#FF6B6B', marker='s', linestyle='--')
plt.plot(df['Mois'], df['2011'], label='2011', color='#6AB04C', marker='x', linestyle='--')
plt.legend()
plt.xticks(x, df['Mois'])
plt.xlabel('Mois')
plt.ylabel('Chiffre de vente')
plt.title('Chiffre de vente par mois pour chaque année entre 2009 et 2011')
plt.show()
```
