```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ocean_dict = {'Pacifique':183.4,'Atlantique':106.7,
              'Indien':73.8,
              'Antarctique':19.7,'Arctique':12.4}
```

```python
ocean_df = pd.DataFrame()

ocean_df['ocean'] = ocean_dict.keys()
ocean_df['superficie'] = ocean_dict.values()
```

```python
ocean_df
```

```python
sns.barplot(y='ocean', x='superficie', data=ocean_df)
plt.title('Superficie des océans en millions de km²')
#plt.xlabel('') # Redondant avec le contexte
plt.ylabel('') # Redondant avec le titre
plt.show()
```
