```python
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
import seaborn as sns
```

```python
continent_dict = {'Asie': 44.58, 'Afrique':30.06, 'Amérique N.' : 24.26, 'Amérique S.': 17.82, 'Antarctique': 13.21, 'Europe': 9.94, 'Océanie': 7.69}
```

```python
continent_df = pd.DataFrame()

continent_df['continent'] = continent_dict.keys()
continent_df['superficie'] = continent_dict.values()

continent_df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='continent', y='superficie', data=continent_df, palette="crest")
plt.title("Superficie des continents du globe terrestre", fontsize=14, weight='bold')
plt.xlabel("Continent")
plt.ylabel("Superficie (en millions de km²)")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(continent_df['superficie'], labels=continent_df['continent'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("crest", len(continent_df)))
plt.title("Répartition de la superficie des continents", fontsize=14, weight='bold')
plt.show()
```

```python
import plotly.express as px

fig = px.treemap(continent_df, path=[px.Constant(""), 'continent'],
                 values='superficie', color='continent',
                 color_discrete_map={'Asie':'yellow',
                                     'Afrique':'brown',
                                     'Amérique N.':'red',
                                     'Amérique S.':'pink',
                                     'Antarctique':'darkblue',
                                     'Europe':'lightgreen',
                                     'Océanie':'orange'})

fig.show()
fig.write_html('ocean.html')
```
