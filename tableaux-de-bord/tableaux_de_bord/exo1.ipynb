```python
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
import seaborn as sns
```

```python
ocean_dict = {'Pacifique':183.4, 'Atlantique':106.7, 'Indien':73.8, 'Antarctique':19.7, 'Arctique':12.4}
```

```python
ocean_df = pd.DataFrame()

ocean_df['ocean'] = ocean_dict.keys()
ocean_df['superficie'] = ocean_dict.values()

ocean_df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='ocean', y='superficie', data=ocean_df, palette="crest")
plt.title("Superficie des océans du globe terrestre", fontsize=14, weight='bold')
plt.xlabel("Océan")
plt.ylabel("Superficie (en millions de km²)")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(ocean_df['superficie'], labels=ocean_df['ocean'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("crest", len(ocean_df)))
plt.title("Répartition de la superficie des océans", fontsize=14, weight='bold')
plt.show()
```

```python
import plotly.express as px

fig = px.treemap(ocean_df, path=[px.Constant(""), 'ocean'],
                 values='superficie', color='ocean',
                 color_discrete_map={'Pacifique':'lightlightblue',
                                     'Atlantique':'lightblue',
                                     'Indien':'aqua',
                                     'Arctique':'blue',
                                     'Antarctique':'darkblue'})

fig.show()
fig.write_html('ocean.html')
```
