```python
import pandas as pd
import matplotlib.pyplot as plt

ocean_dict = {'Pacifique':183.4, 'Atlantique':106.7, 'Indien':73.8, 'Antarctique':19.7, 'Arctique':12.4}

df = pd.DataFrame()

df['Océan'] = ocean_dict.keys()
df['Superficie (millions de km2)'] = ocean_dict.values()

df
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Océan'], df['Superficie (millions de km2)'])
plt.xlabel('Océan')
plt.ylabel('Superficie en millions de km²')
plt.title('Superficie des océans en millions de km²')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.style.use("seaborn-v0_8-colorblind")
plt.title("Répartition de la superficie des Océans")
plt.pie(df['Superficie (millions de km2)'], labels=df['Océan'], autopct="%1.1f%%")
plt.show()
```
